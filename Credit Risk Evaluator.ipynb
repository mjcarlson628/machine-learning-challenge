```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv('Resources/2019loans.csv')
test_df = pd.read_csv('Resources/2020Q1loans.csv')
```

```python
print(len(train_df))
print(len(test_df))
```

```python
set(train_df.columns) == set(test_df.columns)
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_train = pd.get_dummies(train_df.drop(columns = "target"))
X_train
```

```python
y_train = pd.get_dummies(train_df["target"])["high_risk"]
y_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test = pd.get_dummies(test_df.drop(columns = "target"))
X_test
```

```python
y_test = pd.get_dummies(test_df["target"])["high_risk"]
y_test
```

```python
set(X_train.columns) == set(X_test.columns)
```

```python
train_cols = X_train.columns
test_cols = X_test.columns

print(len(train_cols))
print(len(test_cols))
```

```python
for x in train_cols:
    if x not in test_cols:
        print(x)
```

```python
for y in test_cols:
    if y not in train_cols:
        print(y)
```

```python
# add missing dummy variables to testing set
X_test["debt_settlement_flag_Y"] = 0
```

```python
set(X_train.columns) == set(X_test.columns)
```

Random forest should perform better on the unscaled data. The magnitude of values in columns such as "loan_amt" and "annual_inc" might be more likely to distort the importance of these columns in a logistic regression model, when compared with the amount of variation in the categorical columns that have been converted to contain only values of 0 and 1. Random forest should be able to handle the upper range of values more effectively.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
clf = LogisticRegression(max_iter = 5000).fit(X_train, y_train)
print(f"Training Score: {clf.score(X_train, y_train)}")
print(f"Testing Score: {clf.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
rf = RandomForestClassifier(random_state = 1).fit(X_train, y_train)
print(f"Training Score: {rf.score(X_train, y_train)}")
print(f"Testing Score: {rf.score(X_test, y_test)}")
```

```python
# Scale the data
scaler = StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Logistic regression should perform better on scaled data because the variation within each column will be treated on a more comparable level. Having such a large number of columns makes it more likely for random forest to overfit the training data, and scaling is unlikely to change the patterns that led to overfitting the unscaled data.

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf_s = LogisticRegression(max_iter = 5000).fit(X_train_scaled, y_train)
print(f"Training Score: {clf_s.score(X_train_scaled, y_train)}")
print(f"Testing Score: {clf_s.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
rf_s = RandomForestClassifier(random_state = 1).fit(X_train_scaled, y_train)
print(f"Training Score: {rf_s.score(X_train_scaled, y_train)}")
print(f"Testing Score: {rf_s.score(X_test_scaled, y_test)}")
```

Scaling the data led to an improvement from 0.56 to 0.76 in the test score of the logistic regression model, but no improvement in the random forest model score of 0.64. Random forest achieved an accuracy score of 1.0 on the training data, with and without scaling. The test scores of the random forest models are much lower than the training scores, indicating that the model is overfitted.
